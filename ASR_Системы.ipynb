```python
import os

def transcribe_file(path, model):
    result = model.transcribe(path, language="ru")
    return result["text"]

audio_files = [f for f in os.listdir() if f.endswith(".mp3")]
whisper_results = {}

for f in sorted(audio_files):
    print(f"Transcribing {f} …")
    whisper_results[f] = transcribe_file(f, model)
    print(whisper_results[f])
```

```python
from jiwer import wer
import numpy as np

def cer(ref, hyp):
    # простой CER: edit distance на уровне символов / длина
    import editdistance
    return editdistance.eval(ref, hyp) / max(len(ref), 1)

metrics = {}

for fname in manual_transcripts:
    ref = manual_transcripts[fname].lower().strip()
    hyp = whisper_results.get(fname, "").lower().strip()

    word_err = wer(ref, hyp)
    char_err = cer(ref.replace(" ", ""), hyp.replace(" ", ""))

    metrics[fname] = {
        "reference": ref,
        "whisper": hyp,
        "WER": word_err,
        "CER": char_err
    }

import pandas as pd
df = pd.DataFrame.from_dict(metrics, orient="index")
df
```
